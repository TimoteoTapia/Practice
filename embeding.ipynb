```python
import google.generativeai as genai
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.llms.gemini import Gemini
from llama_index.core import Settings
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, VectorIndexRetriever
```

```python
GOOGLE_API_KEY = os.environ["GOOGLE_API_KEY"]
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel(model_name="gemini-1.5-flash")
text_input = "Explain the concept of quantum computing."
response = model.generate_content([text_input])
print(response.text)
```

```python
# Configurar Llama Index
Settings.chunk_size = 700 # Define el tamaño del chunk como desees

# Configurar el modelo y los embeddings
Settings.embed_model = GeminiEmbedding(
    model_name="models/embedding-001"
)
# Configurar el modelo de generación
Settings.llm = Gemini(
    model_name="models/gemini-1.5-flash",  # Asegúrate de usar un modelo que soporte generación
    api_key=GOOGLE_API_KEY
)

# Leer los documentos
files = SimpleDirectoryReader("./data/raw").load_data()

# Crear el índice de vectores
index = VectorStoreIndex.from_documents(files)

print("Índice creado con éxito.")
```

```python
from llama_index.vector_stores.pinecone import PineconeVectorStore
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core.ingestion import IngestionPipeline
```

```python
pinecone_index = pc.Index("hints")
embed_model = GeminiEmbedding(
    model_name="models/embedding-001"
)
vector_store = PineconeVectorStore(pinecone_index=pinecone_index)
pipeline = IngestionPipeline(
    transformations=[SentenceSplitter(chunk_size=700, chunk_overlap=20), embed_model],
    vector_store=vector_store,
)
pipeline.run(documents=files)
```

```python
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.retrievers import VectorIndexRetriever
```

```python
index = VectorStoreIndex.from_vector_store(vector_store=vector_store)
retriever = VectorIndexRetriever(index=index, similarity_top_k=5)
query_engine = RetrieverQueryEngine(retriever=retriever)
```

```python
def format_chat_prompt(message, chat_history):
    prompt = ""
    for turn in chat_history:
        user_message, bot_message = turn
        prompt = f"{prompt}\nUser: {user_message}\nAssistant: {bot_message}"
    prompt = f"{prompt}\nUser: {message}\nAssistant:"
    return prompt

def respond(message, chat_history):

    information = "\n".join(information)

    full_prompt = f"""

    Information:
    START OF INFORMATION

    {information}

    END OF INFORMATION

    {message}
    """

    formatted_prompt = format_chat_prompt(full_prompt, chat_history)
 
    chat_history.append((message, bot_message))
    return "", chat_history

with gr.Blocks() as demo:
    chatbot = gr.Chatbot(height=480)  # Ajusta el tamaño según sea necesario
    msg = gr.Textbox(label="Question")
    btn = gr.Button("Submit")
    clear = gr.ClearButton(components=[msg, chatbot], value="Clear console")

    btn.click(respond, inputs=[msg, chatbot], outputs=[msg, chatbot])
    msg.submit(respond, inputs=[msg, chatbot], outputs=[msg, chatbot])  # Presiona Enter para enviar

demo.launch(share=True)
```
